```python
from collections import deque
def bfs(graph,start):
    visited=set()
    queue=deque([start])
    visited.add(start)
    while queue:
        vertex=queue.popleft()
        print(vertex,end="")
        for neigbhor in graph[vertex]:
            if neigbhor not in visited:
                visited.add(neigbhor)
                queue.append(neigbhor)
graph={
        'B':['D','E'],
        'D':['G'],
    'E':['C'],
        'G':['H'],
        'H':['K'],
        'F':[]            
}
print("BFS Traversal")
bfs(graph,"B")
```
