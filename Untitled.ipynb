##Pandas Methods

```python
import pandas as pd
data_frame=pd.DataFrame({
    'Name':['Alice','Bob','Chicago'],
    'Age':[25,30,35],
    'city':['New York','Los Angles','Chicago']
})
print('1)describe() method')
print(data_frame.describe())
print('2)head() method')
print(data_frame.head(2))
print('3)tail() metho')
print(data_frame.tail(2))
print('4)info() method')
print(data_frame.info())
```

##Pandas_Vizaulization

```python
import pandas as pd
import matplotlib.pyplot as plt
data ={
    'Subject':['Maths','Physics','Chemistry','English','Cs'],
    'Marks':[78,85,69,74,90]
}
df=pd.DataFrame(data)
#Bar Plot
df.plot(x='Subject',y='Marks',kind='bar',title='Bar Plot',color='r')
plt.show()
#Histogram
df['Marks'].plot(kind='hist',title='Histogram',color='y')
plt.show()
#Line plot
df.plot(x='Subject',y='Marks',kind='line',title='Line Plot',color='g')
plt.show()
#Scatter plot
df.plot(x='Subject',y='Marks',kind='scatter',title='Scatter Plot',color='b')
plt.show()
```
